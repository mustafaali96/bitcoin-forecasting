```python
import requests
import matplotlib.pyplot as plt

price = []
time = []
coin1 = ""
coin2 = ""

def getDatafromURL(coinName,convertTo,datapoints):
    url = "https://min-api.cryptocompare.com/data/histominute?" \
          "fsym="+coinName.upper()+"&tsym="+convertTo.upper()+"&limit="+datapoints+"&aggregate=3&e=CCCAGG"
    
    global coin1
    coin1 = coinName
    global coin2
    coin2 = convertTo
    
    req = requests.get(url)
    Fulldata = req.json()['Data']
    for data in Fulldata:
        time.append(float(data['time']))
        price.append(float(data['close']))

def plotting():
    plt.plot(time, price)
    plt.title(coin1.upper() + ' convert to '+ coin2.upper())
    plt.xlabel('Time in Seconds')
    plt.ylabel('Price in '+ coin2.upper())
    plt.show()
```

```python
getDatafromURL("btc","usd","99")
```

```python
plotting()
```
